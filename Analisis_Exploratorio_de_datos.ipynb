# Analisis Exploratorio de datos
## Univariado, Bivariado, Multivariado

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el archivo EXCEL
df = pd.read_excel("https://github.com/restevesd/DataCleaning/blob/main/limpiezatransformacion.xlsx?raw=true")
```

```python
# Descripción estadística de los datos
df.describe()
```

```python
# Visualización de la distribución de la edad
sns.histplot(data=df, x='edad')
plt.show()
```

```python
# Visualización de la relación entre ingreso_neto y gasto_tarjeta
sns.scatterplot(data=df, x='ingreso_neto', y='gasto_tarjeta')
plt.show()
```

```python
# Visualización de la relación entre edad y gasto_tarjeta
sns.scatterplot(data=df, x='edad', y='ingreso_neto')
plt.show()
```

```python
sns.scatterplot(data=df, x='edad', y='ingreso_neto', hue="codigo_genero")
```

```python
#Pairplot para las variables edad, ingreso_neto y gasto_tarjeta
sns.pairplot(data=df,vars=['edad', 'ingreso_neto', 'gasto_tarjeta'], hue='codigo_genero')
plt.show()
```

```python
# Identificación de valores atípicos en la columna ingreso_neto
sns.boxplot(data=df, x='ingreso_neto')
plt.show()
```

```python
# Identificación de valores atípicos en la columna gasto_tarjeta
sns.boxplot(data=df, x='gasto_tarjeta')
plt.show()
```
