```python
url = 'https://www.resultados-futbol.com/primera2024/grupo1/jugadores'
```

```python
import requests
import json
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
def write_doc_json(content, mode):
    own_file = open(path_file_dest, mode, encoding='iso-8859-1')
    own_file.write(content.decode('utf-8'))
    own_file.close()
    print('\n•  El archivo json fue sobre-escrito con exito!')
```

```python
path_file_dest = r'futbol.json'
header = {'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'}
response = requests.get(url, headers=header)
response.text[0:300]
```

```python
if response.status_code == 200:

    info = []
    soup = BeautifulSoup(response.text, "html.parser")
    equipos = soup.findAll('div', attrs={'class': 'boxhome boxhome-2col'})

    for x in range (0, 5): #len(equipos)):
        # Obtengo nombre del equipo
        nombre_equipo = equipos[x].find('span', {'class': 'titlebox'}).get_text().strip()
        # Obtengo la tabla de jugadores
        tabla = equipos[x].find('tbody')
        # Obtengo todos los jugadores en una lista con su arbol HTML
        jugadores = tabla.findAll('tr')

        print('________________________________________________________________')
        print('')
        print(nombre_equipo)
        print('')

        for i in range (0, len(jugadores)):

            datos_jugador = []

            # Ingreso por índice y busco la etiqueta "td" y la etiqueta "strong"
            nombre = jugadores[i].findAll('td')[0].find('strong').get_text()
            url_jugador = 'https://www.resultados-futbol.com' + jugadores[i].find('td').find('a')['href']

            # Obtengo HTML con datos del JUGADOR
            response_dos = requests.get(url_jugador, headers=header)

            print(nombre)

            if response_dos.status_code == 200:
                info_dos = []
                soup_dos = BeautifulSoup(response_dos.text, "html.parser")
                data_dos = soup_dos.findAll('dd')
                #pprint(data_dos[0:9])

                try:
                    for l in range (0, 9):

                        datos_jugador.append(data_dos[l].get_text().strip())

                    pprint(datos_jugador)

                    individuo = {
                        'Equipo' : nombre_equipo,
                        'Nombre_Completo' : datos_jugador[1],
                        'Edad' : datos_jugador[2],
                        'Fecha_Nacimiento' : datos_jugador[3],
                        'Provincia' : datos_jugador[4],
                        'Pais' : datos_jugador[5],
                        'Nacionalidad' : datos_jugador[6],
                        'Posicion' : datos_jugador[7],
                        'Altura' : datos_jugador[8]
                        }

                    info.append(individuo)
                except:
                    pass
        print('')

    content = json.dumps(info, indent=4, sort_keys=True, ensure_ascii=False).encode('utf-16')
    write_doc_json(content ,'w+')
    #data = pd.read_json(path_file_dest, encoding = 'UTF-8')
    #print('•  Se leyo el json y se creo el dataframe data')

else:
    print('No se pudo conectar')
```
